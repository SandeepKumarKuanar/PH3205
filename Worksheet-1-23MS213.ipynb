```python
###  Q1
#function to check if the given integer is an prime or not
def is_prime(x):
    if x < 0: 
        return 0
    
    root_number = int((x**0.5) // 1)
    is_it = 1
    
    for i in range(2, root_number + 1):
        if x % i == 0:
            is_it = 0
            break

    return is_it 

check_number = int(input("Give me a number:\n").strip())
if is_prime(check_number):
    print("This is a prime number")
else:
    print("This is not a prime number")
```

```python
###  Q2
#all possible primes between 2 numbers 
num1 = int(input("Give me the first number:\n").strip())
num2 = int(input("Give me the second number:\n").strip())

list_of_primes = [i for i in range(num1 + 1, num2) if is_prime(i) == 1]
print(list_of_primes)
```

```python
###  Q3
# that evaluates the series exp(x) 
def f(x, N):
    current_value = 1
    upper = 1
    lower = 1
    for i in range(1, N):
        upper *= x
        lower *= i
        value = (upper / lower)
        current_value += value
    return current_value

x = float(input("Give me the x:\n").strip())
N = int(input("Give me the number of terms:\n").strip())
print(f(x, N))
```

```python
###  Q5
## defining the leibniz method to find the pi's value
import numpy as np
import matplotlib.pyplot as plt

def leibniz(N):
    value = 0
    for i in range(0, N):
        lower = (2 * i) + 1
        if i % 2 == 0:
            value += 1 / lower
        else:
            value += -1 * (1 / lower)
    pi = 4 * value
    return pi

def continued_fraction(N):
    value = 0
    for i in range(2*N - 1, 0, -2):
        value = (i**2) / (2 + value)
    
    pi = 4 / (value + 1)
    return pi
    
list_from_leibniz = [leibniz(i) for i in range(1, 11)]
list_from_continued_fraction = [continued_fraction(i) for i in range(1, 11)]
x = [i for i in range(1, 11)]

print("From Leibniz's method")
for i in range(0, len(list_from_leibniz)):
    print(f"{list_from_leibniz[i]:.8f}", end = " ")

print("\n")
print("From Continued fraction's method")
for i in range(0, len(list_from_continued_fraction)):
    print(f"{list_from_continued_fraction[i]:.8f}", end = " ")
    
y1 = np.array(list_from_leibniz)
y2 = np.array(list_from_continued_fraction)
print("\n")
#-------Plotting--------
# Create the plot
plt.figure(figsize=(10, 6))

plt.plot(x, y1, label='The Leibniz\'s method', color='blue')
plt.plot(x, y2, label='The Continued fraction\'s method', color='red', linestyle='--')

# Add labels and a legend
plt.xlabel('X-axis values')
plt.ylabel('Y-axis values (same units)')
plt.title('Comparing the convergence')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
