### Importing the 2 basic requirements below

```python
import numpy as np
import matplotlib.pyplot as plt
```

### setting the error permitted and the maximum times a while loop can run

```python
esp = 1.0e-5
maxiter = 50
```

## Question 1
- Analysing the intervals of the functions 
- taking proper ranges and then fixing the guesses

```python
def f(x):
    return (3*x - np.tan(x))

def f_dash(x):
    return (3 - (1 / np.cos(x))**2)

#-------Plotting--------
# Create the plot
plt.figure(figsize=(10, 6))
x = np.linspace(0, 1.57, 1000)
y = f(x)
plt.plot(x, y, label='Given function', color='red')
plt.ylim(-3, 10) ## limiting how much we see in y-axis
# Add labels and a legend
plt.xlabel('X-axis input values')
plt.ylabel('Y-axis function values')
plt.title('Plotting and testing the function out')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```

## Q1e

```python
def bisection(f, left, right, error, maxiter):
    if f(left) * f(right) >= 0:
        print(f"Bad interval given, choose a better interval than [{left}, {right}]")
        return None
    else:
        midpoint = (left + right) / 2
        iter = 0
        
        while iter <= maxiter and abs(f(midpoint)) > error:
            #print(f"{left}, {right}, {iter}")
            # we redefine midpoints
            if f(left) * f(midpoint) < 0:
                right = midpoint
                midpoint = (left + right) / 2
            else:
                left = midpoint
                midpoint = (left + right) / 2
            iter += 1
        
    return midpoint, iter

root, steps = bisection(f, 1.2, 1.4, esp, maxiter)
print(f"The root for the function = {root} and in {steps} steps using bisection method")
```

## Q1f

```python
def secant(f, x1, x2, error, maxiter):
    intersection = x2 - f(x2) * ((x2 - x1)/ (f(x2) - f(x1)))
    iter = 0

    while iter <= maxiter and abs(f(intersection)) > error:
        #print(f"{x1}, {x2}, {intersection}, {iter}")
        intersection = 0
        intersection = x2 - f(x2) * ((x2 - x1)/ (f(x2) - f(x1)))
        x2, x1 = intersection, x2
        iter += 1

    return intersection, iter

root, steps = secant(f, 1.2, 1.25, esp, maxiter)
print(f"The root for the function = {root} and in {steps} steps using secant method")
```

## Q1g

```python
def newton_raphson(f, x1, error, maxiter):
    x2 = x1 - (f(x1) / f_dash(x1))
    iter = 0

    while iter <= maxiter and abs(f(x2)) > error:
        #print(f"{x1}, {x2}, {intersection}, {iter}")
        x2 = x1 - (f(x1) / f_dash(x1))
        x1 = x2
        iter += 1

    return x2, iter

root, steps = newton_raphson(f, 1.2, esp, maxiter)
print(f"The root for the function = {root} and in {steps} steps using newton raphson method")
```

## Question 2
- Analysing the intervals of the V(x) given
- Finding out Kinetic energies and plotting it as well
- taking proper ranges and then fixing the guesses

```python
def V(x):
    return x**4 + ((2*x) / 3)**3 - x**2

TE = -0.125 #in electron-volts 


#-------Plotting--------
# Create the plot
plt.figure(figsize=(10, 6))
x = np.linspace(-1.25, 1.25, 1000)
y = V(x)
plt.plot(x, y, label='V(x)', color='blue')
plt.axhline(y = 0, color='red', linestyle='--', label='y = 0 line')
plt.axhline(y = TE, color='black', label='y = 0 line')
# Add labels and a legend
plt.xlabel('X-axis input values')
plt.ylabel('Y-axis function values')
plt.title('Plotting and testing the voltage out')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```

```python
def KE(x):
    return TE - V(x)

#-------Plotting--------
# Create the plot
plt.figure(figsize=(10, 6))
x = np.linspace(-1.25, 1.25, 1000)
y = KE(x)
plt.plot(x, y, label='V(x)', color='blue')
plt.axhline(y = 0, color='red', linestyle='--', label='y = 0 line')
# Add labels and a legend
plt.xlabel('X-axis input values')
plt.ylabel('Y-axis function values')
plt.title('Plotting and testing the voltage out')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```

### 2a

```python
#### there exists 4 roots to this equation as we can see
### using the Bisection method only
## first root
root1, steps1 = bisection(KE, -1.25, -1, esp, maxiter)
print(f"First root is found {root1}, and found after {steps1}")
## second root
root2, steps2 = bisection(KE, -0.5, -0.25, esp, maxiter)
print(f"Second root is found {root2}, and found after {steps2}")
## third root
root3, steps3 = bisection(KE, 0.25, 0.5, esp, maxiter)
print(f"Third root is found {root3}, and found after {steps3}")
## fourth root
root4, steps4 = bisection(KE, 0.55, 0.8, esp, maxiter)
print(f"Fourth root is found {root4}, and found after {steps4}")
```
